I've explain the concept of training data and test data using a simple analogy:

Training Data is like a student practicing math problems from a textbook before an exam. The student learns from the textbook, solves various math problems, and understands different problem-solving techniques. This is the data that the student "trains" on, just like a machine learning model "learns" from training data.

Test Data is like the actual exam. During the exam, the student faces new, unseen math questions that are not from the textbook. The student uses the knowledge gained from practicing on the training data (the textbook) to solve these new questions and predict the answers.

In the context of machine learning:

Training Data is the part of the dataset used to teach the machine learning model. It's like the textbook that the model learns from.
Test Data is the part of the dataset that the model hasn't seen during training. It's like the new exam questions that the model needs to answer based on what it learned from the training data.
The goal of training and testing is to ensure that the machine learning model can perform well not just on the problems it has already seen (training data, like textbook exercises) but also on new, unseen problems (test data, like exam questions). This helps evaluate how well the model can generalize its learning to real-world situations.

Model Evaluation
   
       Accuracy Score

```python
# Accuracy on training data
X_train_predication = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predication,Y_train)
```

    Sure, here's a simplified explanation:

    1. X_train_predication = model.predict(X_train) :
    The model makes predictions on the training data and stores them in ' X_train_predication ' .

    2. training_data_accuracy = accuracy_score(X_train_predication, Y_train) "
    This line calculates how accurate the model's predictions are when compared to the actual correct answers in the training data.
    The result, ' training_data_accuracy ',
    represents the percentage of correct predictions made by the model on the training data.
    It's a measure of how well the model has learned from the data it was trained on.

```python
print('Accuracy on Training Data :',training_data_accuracy)
```

```python
 # accuracy on test data
 X_test_predication = model.predict(X_test)
 test_data_accuracy = accuracy_score(X_test_predication,Y_test)
```

# 01. task

     EMAIL SPAM DETECTION

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#df is the dataframe
df = pd.read_csv("spam.csv",encoding = "ISO-8859-1")
df.head()
```

```python
#finding null values
df.isnull().sum()
```

```python
#dropping null values
df1 = df.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"], axis=1)
```

```python
df1.head()
```

```python
df1.rename(columns = {"v1" : "Spam or Ham", "v2":"Mail Message"},inplace = True)
df1.head()
```

```python
df1.shape
```

```python
#Converting the categorical values into numerical values
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df1['Spam or Ham'] = encoder.fit_transform(df1['Spam or Ham'])
```

```python
#Finding duplicates
df1.duplicated().sum()
```

```python
#dropping duplicates

df1 = df1.drop_duplicates(keep = 'first')
```

```python
df1.duplicated().sum()
```

```python
#Splitting the data
X =  df1["Mail Message"]
y =df1["Spam or Ham"]
```

```python
#Splitting the data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.20, random_state = 0)
```

```python
# Convert the text data into a matrix of token counts
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_train_count = cv.fit_transform(X_train.values)
X_train_count.toarray()
```

```python
#Fitting multinomial naive bayes
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
from sklearn.metrics import confusion_matrix , recall_score , precision_score
from sklearn.metrics import accuracy_score
```

```python
#Testing  the mail (spam/ham)

mail_ham = ['Same. Wana plan a trip sometme then']
mail_ham_count = cv.transform(mail_ham)
y_pred = model.predict(mail_ham_count)
y_pred
```

```python
#finding accuracy of the training dataset
model.score(X_train_count,y_train)
```

```python
#finding accuracy of the test dataset
X_test_count = cv.transform(X_test)
model.score(X_test_count,y_test)
```
